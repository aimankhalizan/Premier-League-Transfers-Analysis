```python
#importing panda library 
import pandas as pd

```

Importing the necessary libraries is key

```python
#Importing the csv file into pandas
df = pd.read_csv(r'C:\Users\Windows 10\OneDrive - Universiti Teknologi MARA\Desktop\EPL Standings 2000-2022.csv', index_col = False)
df
```

```python
#Counting the duplicates
df.duplicated().sum()
```

First thing we need to do is make sure that there are no duplicates in the dataset. If there are, the duplicates are to be removed. 

```python
#Select only the last five seasons 
df = df[df['Season'].isin(['2018-19','2019-20','2020-21','2021-22'])]
df
```

Because there are a lot of data involving football transfers which dates back to the year 2000, I have decided to only take rows from the past 5 years. 

```python
#Reset the index so we can have a cleaner primary key later on 
df = df.reset_index()
df
```

```python
#drop column named index
df = df.drop(['index'], axis = 1)
df
```

```python
#Renaming 'season' column to match with the 'season' column in the 'transfers 
df['Season'] = df['Season'].replace(['2018-19','2019-20','2020-21','2021-22'], ['2018/2019','2019/2020','2020/2021','2021/2022'])
df
```

The season format is changed to match with the season format in the Transfers table. This is so that I can join these two tables together when the data is imported into MSSQL later.

```python
#Renaming rows in 'Qualification or relegation' column to a simpler name 
df['Qualification or relegation'] = df['Qualification or relegation'].replace(['Qualification for the Champions League group stage','Relegation to the EFL Championship'],['Champions League','Relegated'])
df
```

Renamed the values to a more easy-to-read and simpler value.

```python
#Renaming columns 
df= df.rename(columns = {'Season':'season','Pos':'position','Team':'club_name','Pld':'played','Pts':'points','Qualification or relegation':'qualification/relegation'})
df
```

These columns are also renamed to make them more readable and understandable to readers. 

```python
#Renaming the index column
df.index.name = 'standings_id'
df
```

The Standings ID will be the primary key for this particular table. 

```python
#Export the dataframe into a csv file
df.to_csv('standings_cleaned.csv',index = True)
```
