```python
#import libraries
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
from importlib import reload
```

```python
#importing the datasets 
df_transfer = pd.read_csv('transfers_cleaned.csv')
df_sold = pd.read_csv('transfers_out.csv')
df_totalSpent_avgFinish = pd.read_csv('total_spent_average_finish.csv')
df_position_total_signed = pd.read_csv('position_total_signed.csv')
```

```python
#Finding out the average spending for each teams
average_spend = df_transfer.groupby('club_name')['transfer_fee(millions)'].mean()
print(average_spend)
```

From here, we can see that Chelsea have the highest average spending out of all teams in the premier league. 

```python
#Plotting a scatter graph to find the correlation between Premier League team's spending and success
total_spent = df_totalSpent_avgFinish['total_spent']
average_finish = df_totalSpent_avgFinish['average_finish']

plt=reload(plt)
plt.scatter(total_spent,average_finish,color = 'blue', edgecolors='black')
plt.xlabel('Total Spent (Millions)')
plt.ylabel('Average Finish')
plt.title('Scatter plot of teams average finish and their total spending')

plt.show()
```

This graph is plotted so that we can see if a team's success is related to their spending. The y-axis is the final position in the Premier League table of teams. The higher the number on the y-axis, the less successful the team is. While the x-axis refers to the total spending of each team. From the scatter plot diagram, we can see that the higher the spending is, the more succesful they are. While teams with less spending are also less succesful.

```python
#Plotting a bar chart to find out the most in-demand player position 
player_position = df_position_total_signed['position']
total_signed = df_position_total_signed['total_signed']


plt=reload(plt)
plt.barh(player_position,total_signed, color = 'green', edgecolor = 'black')
plt.xlabel('Total Signed')
plt.ylabel('Player Position')
plt.title('Graph of Total Signed by each position')

plt.show()
```

This graph is plotted to find out the most bought position in the Premier League. The y-axis is the player positions, while the x-axis the the number of players bought for each position. From the graph, we can see that center-backs are the most in-demand position in the Premier League. This could be because recently, there is a change in trend of having center-backs that are great at passing, in which is a skill that older center-backs lack. 

```python
#Plotting a bar chart to find out the average spending fee for each position in-top 4 teams
df1 = pd.read_csv('average_top4_spent.csv')
position = df1['position']
average_spent= df1['average_spent']

plt=reload(plt)
plt.barh(position,average_spent, color = 'blue', edgecolor = 'black')
plt.xlabel('Average Fee')
plt.ylabel('Player Position')
plt.title('Graph of average fee for each position for top 4 teams')

plt.show()
```

From here, we can see that the highest fee is the attacking midfield position, followed by the defensive midfield.
```python
#Displaying the frequency of players age 
df = pd.read_csv('age_distribution.csv')
df
```

```python
#Displaying the above table in a graph format 
age = df['age']
frequency = df['frequency']

plt.figure().set_figwidth(15)
plt.bar(age,frequency, color = 'purple', edgecolor = 'black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Graph of players age and the frequency')
plt.show()
```

This bar diagram explains the frequency of signing for each age. We can see that 23 has the highest frequency compared to other ages. There are also higher frequencies for players aged between 19-29. This could be because teams are looking to sign players at a younger age because they are considered better investments compared to older players. 

```python
#Plotting a scatter graph to display the average age for each position signed
df = pd.read_csv('position_age.csv')

position = df['position']
age = df['average_age']

plt.scatter(age,position, color = 'navy', edgecolors = 'black')
plt.xlabel('Age')
plt.ylabel('Position')
plt.title('Average age by position')

plt.show()
```

This diagram shows the average age for each position. Most position ages are in between 23-25. However, the goalkeeper is the outlier position. This is because most goalkeepers tend to be at their best when they are at their 30's. Goalkeepers also tend to only retire when they are at their late 30's or early 40's. Therefore, teams are not afraid to sign goalkeepers that are in their 30's. 

From this analysis, we can conclude that there are correlation between a team's spending and their success. We also found out that center-backs are the most sought after position in the Premier League and they do not cost as much as attacking players. Most players are also bought when they are in their early 20's. Therefore, the most profitable signing that we can make is to buy a young attacking player at an age below 20, and then we resell them when they reach their peak value at the age of early 20s to make the most profit.
